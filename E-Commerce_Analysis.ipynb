```python
def plot(df , figsize , subplot):
    
    plt.figure(figsize = figsize)
    for index , column in enumerate(df.columns):
        plt.subplot(*subplot , index+1)
        sns.histplot(df[column])
        plt.xticks(rotation = 'vertical')
```

# Analyzing pricing trends and discount strategies

```python
#Analyzing pricing trends and discount strategies
price_stats = df1['actual_price'].describe()

# Display the statistics
print(price_stats)
```

```python
#Visualize the Distribution of Prices
# Histogram of Actual Prices
plt.figure(figsize=(10, 6))
sns.histplot(df1['actual_price'], bins=30, kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Frequency')
plt.show()
```

```python
#Visualize the Distribution of Prices
# Histogram of Actual Prices
plt.figure(figsize=(10, 6))
sns.histplot(df1['selling_price'], bins=30, kde=True)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Prices')
plt.ylabel('Frequency')
plt.show()
```

```python
#Correlation Between Discounts and Sellingprice:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discount', y='selling_price', data=df1)
plt.title('Scatter Plot of Discounts vs. Selling Prices')
plt.xlabel('Discount')
plt.ylabel('Selling Price')
plt.show()
```

```python
#Correlation Between Discounts and actual price:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discount', y='actual_price', data=df1)
plt.title('Scatter Plot of Discounts vs. Actual Prices')
plt.xlabel('Discount')
plt.ylabel('Actual Price')
plt.show()
```

```python
# Average discount percentages
df1['discount_percentage'] = (df1['actual_price'].astype(float) - df1['selling_price'].astype(float)) / df1['actual_price'].astype(float) * 100
average_discount = df1['discount_percentage'].mean()
print(f'Average Discount Percentage: {average_discount:.2f}%')
```

```python
# Explore discount patterns by category
plt.figure(figsize=(12, 6))
sns.barplot(x='category', y='discount_percentage', data=df1)
plt.title('Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Discount Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate average discount for each category
seller_discounts = df1.groupby('category')['discount'].mean().reset_index()

# Select top 10 category based on average discount
top_10_category = seller_discounts.nlargest(10, 'discount')

# Plot top 10 category vs discount
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_category, x='category', y='discount', palette='viridis')
plt.title('Top 10 category vs Discount')
plt.xlabel('category')
plt.ylabel('Discount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
     
```

```python
# Select the top 20 brands for visualization
top_brands = df1['brand'].value_counts().head(10).index

# Filter data for top brands
top_brands_data = df1[df1['brand'].isin(top_brands)]
# Explore discount patterns by the top 10 brands
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='discount_percentage', data=top_brands_data, order=top_brands)
plt.title('Discount Percentage by Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Discount Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = df1[['actual_price', 'selling_price', 'discount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Customer preferences and product satisfaction

```python
# Distribution of Ratings
plt.figure(figsize=(30, 10))
sns.countplot(x='average_rating', data=df1)
plt.title('Distribution of Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Customer satisfaction analysis
#Average Rating vs. Category (Bar Plot)
plt.figure(figsize=(14, 8))
sns.barplot(x='category', y='average_rating', data=df1)
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average rating for each brand
brand_ratings = df1.groupby('brand')['average_rating'].mean().reset_index()
# Select top 10 brands based on average rating
top_10_brands_rating = brand_ratings.nlargest(10, 'average_rating')
# Plot top 10 brands vs average rating
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_brands_rating, x='brand', y='average_rating', palette='viridis')
plt.title('Top 10 Brands vs Average Rating')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate average rating for each seller
seller_ratings = df1.groupby('seller')['average_rating'].mean().reset_index()
# Select top 10 sellers based on average rating
top_10_sellers_rating = seller_ratings.nlargest(10, 'average_rating')
# Plot top 10 sellers vs average rating
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_sellers_rating, x='seller', y='average_rating', palette='viridis')
plt.title('Top 10 Sellers vs Average Rating')
plt.xlabel('Seller')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
